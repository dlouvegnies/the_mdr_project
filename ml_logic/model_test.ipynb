```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
import string
from nltk.tokenize import word_tokenize
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.neighbors import NearestNeighbors
```

```python
news_df = pd.read_csv("../raw_data/export_2024-03-18.csv")
news_df = pd.DataFrame(news_df["title"])
news_df = news_df.drop_duplicates()
```

```python
 # Instantiating the TfidfVectorizer
tf_idf_vectorizer = TfidfVectorizer(max_features=100)

test_list = []
for i in range(0,200000):
    titre = news_df.iloc[i]['title']
    titre = np.str_(titre)
    test_list.append(titre)

weighted_words = pd.DataFrame(tf_idf_vectorizer.fit_transform(test_list).toarray(),columns = tf_idf_vectorizer.get_feature_names_out())
weighted_words
```

```python
# Transformer le DataFrame en une liste de listes
samples = weighted_words.values.tolist()
neigh = NearestNeighbors(n_neighbors=1)
#Fit
neigh.fit(samples)
```

```python
indice = 6532
neigh_dist, neigh_ind = neigh.kneighbors([samples[indice]],10)
```

```python
print("ORIGINE :" +news_df.iloc[indice]['title'])
for i in neigh_ind[0]:
    print(str(i) + "--"+news_df.iloc[i]['title'])
```
